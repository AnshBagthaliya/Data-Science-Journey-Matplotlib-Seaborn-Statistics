```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.normal(0,100,100)
```

```python
df = pd.DataFrame({'x':data})
df['x'].skew()
```

```python
sns.histplot(x='x',data=df)
plt.show()
```
