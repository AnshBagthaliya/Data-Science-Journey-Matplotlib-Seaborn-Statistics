```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('tips.csv')
```

```python
dataset['sex'].mode()
```

```python
md = np.median(dataset['total_bill'])
```

```python
sns.histplot(x='total_bill',data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)], color='red')
plt.plot([md for i in range(0,300)],[i for i in range(0,300)], color='b')
plt.show()
```
