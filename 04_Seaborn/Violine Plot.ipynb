# Violine Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset('tips')
var
```

```python
sns.violinplot(x='day', y='total_bill',data=var)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill',data=var,hue='time')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill',data=var,linewidth = 5,palette="Dark2",saturation=0.4)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill',data=var,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill',data=var,color="red")
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill',data=var,hue='sex',split=True,scale="width")
plt.show()
```

```python
sns.violinplot(x='total_bill' , y='day',data=var,hue='sex')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill',data=var,inner='quart'#('point','stick','box',None))
plt.show()
```

```python
sns.violinplot(x=var['total_bill'])
plt.show()
```
