```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
df = sns.load_dataset('anagrams')
x= df.drop(columns=['attnr'],axis=1).head(10)
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(var,vmin=0,vmax=20)
plt.show()
```

```python
sns.heatmap(var,cmap='PuOr')
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True)
plt.show()
```

```python
var2 = np.linspace(1,10,10).reshape(2,5)
var2
```

```python
sns.heatmap(var2,vmin=0,vmax=10,cmap='PuOr',annot=True)
plt.show()
```

```python
ar = np.array([['a0','a1','a2','a3','a4'],['b0','b1','b2','b3','b4']])
ar
```

```python
sns.heatmap(var2,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s')
plt.show()
```

```python
y={'fontsize':20,'color':'r'}
sns.heatmap(var2,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y)
plt.show()
```

```python
sns.heatmap(var2,vmin=0,vmax=10,cmap='PuOr',annot=True,linewidth=4)
plt.show()
```

```python
sns.heatmap(var2,vmin=0,vmax=10,cmap='PuOr',annot=True,linewidth=4,linecolor='r')
plt.show()
```

```python
sns.heatmap(var2,vmin=0,vmax=10,cmap='PuOr',annot=True,cbar=False)
plt.show()
```

```python
v = sns.heatmap(var2,vmin=0,vmax=10,cmap='PuOr',annot=True,cbar=False)
v.set(xlabel='Python',ylabel='AB07')
sns.set(font_scale=1)
plt.show()
```
