# Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset('tips')
var
```

```python
sns.barplot(x=var.sex,y = var.tip)
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=var,hue='day')
plt.show()
```

```python
ord = ['Female','Male']
sns.barplot(x='sex',y='tip',data=var,hue='day',order = ord)
plt.show()
```

```python
ord = ['Female','Male']
sns.barplot(x='sex',y='tip',data=var,hue='day',order = ord,hue_order = ['Sun','Thur','Fri','Sat'])
plt.show()
```

```python
ord = ['Female','Male']
sns.barplot(x='sex',y='tip',data=var,hue='day',order = ord,hue_order = ['Sun','Thur','Fri','Sat'],ci=40)
plt.show()
```

```python
ord = ['Female','Male']
sns.barplot(x='sex',y='tip',data=var,hue='day',order = ord,hue_order = ['Sun','Thur','Fri','Sat'],orient='h')
plt.show()
```

```python
sns.barplot(x='total_bill',y='tip',data=var,orient = 'h')
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=var,color='r')
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=var,hue='day',palette='icefire')
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=var,hue='day',saturation=50)
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=var,hue='day',errcolor='gold',errwidth=7)
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=var,hue='day',errcolor='y',capsize=0.5)
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=var,hue='day',saturation=50,dodge=False)
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=var,hue='day',saturation=100,alpha=0.5)
plt.show()
```

```python
sns.set(style='whitegrid')
sns.barplot(x='sex',y='tip',data=var,hue='day',)
plt.show()
```
