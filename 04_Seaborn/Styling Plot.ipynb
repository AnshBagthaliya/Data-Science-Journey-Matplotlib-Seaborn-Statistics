# Styling Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset('tips')
var
```

```python
sns.boxplot(y=var['total_bill'])
plt.show()
```

```python
sns.set_style('darkgrid')
sns.barplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,5))
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.show()
```

```python
sns.set_style('white')
sns.set_context('poster',font_scale=1) # talk,paper
sns.barplot(x='day',y='total_bill',data=var,palette='inferno')
plt.show()
```
