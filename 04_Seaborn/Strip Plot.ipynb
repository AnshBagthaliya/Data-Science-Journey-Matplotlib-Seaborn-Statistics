# Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset('tips')
var
```

```python
sns.stripplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue="sex")
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue="sex",palette='rocket_r')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue="sex",palette='rocket_r',linewidth=3,edgecolor='r')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue="sex",palette='rocket_r',jitter=5)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue="sex",palette='rocket_r',size=10,marker='<')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=var,hue="sex",palette='rocket_r',size=10,marker='<',alpha=0.2)
plt.show()
```

```python
sns.stripplot(x=var['total_bill'])
plt.show()
```
