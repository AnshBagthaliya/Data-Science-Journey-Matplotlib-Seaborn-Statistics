# Line PLot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7]
var_2  = [2,3,4,1,5,6,7]

x = pd.DataFrame({'var':var,'var_2':var_2})

sns.lineplot(x='var', y='var_2',data=x)
plt.show()
```

```python
plt.plot(var,var_2)
plt.show()
```

```python
Y_1 = sns.load_dataset("titanic").head(20)
Y_1
```

```python
sns.lineplot(x='age', y='fare',data=Y_1,hue = 'sex',size=50,style="sex",
             palette='Accent',markers=["o",">"],dashes=False,legend=False)
plt.grid()
plt.title('python')
plt.show()
```
