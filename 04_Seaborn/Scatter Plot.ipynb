# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset('penguins').head(20)
var
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var)
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex',size='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',style='sex',size='sex',sizes=(80,20))
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',palette='gist_ncar')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=var,hue='sex',palette='gist_ncar',alpha=0.2)
plt.show()
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=var,
    hue='sex',
    style='sex' ,        #  this tells seaborn to use different markers
    markers=['D', 'X']      # optional: defines which markers to use
)
plt.show()
```
