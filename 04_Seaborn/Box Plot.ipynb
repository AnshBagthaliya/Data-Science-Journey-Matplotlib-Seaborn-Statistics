# Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',data=var,hue='sex')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',data=var,hue='sex',color='g')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',data=var,hue='sex',order=['Fri','Sun','Thur','Sat'])
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',data=var,color='g',
            showmeans=True,meanprops={'marker':'+',"markeredgecolor":'b'})
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',data=var,
            showmeans=True,meanprops={'marker':'+',"markeredgecolor":'b'},linewidth=3,palette='plasma')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='total_bill',y='day',data=var,
            showmeans=True,meanprops={'marker':'+',"markeredgecolor":'b'},linewidth=3,palette='plasma')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='total_bill',y='day',data=var,
            showmeans=True,meanprops={'marker':'+',"markeredgecolor":'b'},linewidth=3,palette='plasma')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,
            showmeans=True,meanprops={'marker':'+',"markeredgecolor":'b'},linewidth=3,palette='plasma',orient='h')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=var['total_bill'])
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=var['total_bill'])
plt.show()
```
