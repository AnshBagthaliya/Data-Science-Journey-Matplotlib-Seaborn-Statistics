```python
#plt.subpot(nrows,ncolumn,index)

import matplotlib.pyplot as plt

x  = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1) # 1 row,2 column,1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) # 1row, 2column, 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart ')

plt.show()
```

```python
import matplotlib.pyplot as plt

#fig, ax = plt.subplot(nrows,ncols,figsize=(width,hight))
fig, ax = plt.subplots(1,2,figsize = (10,5))

x  = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

#fig, ax = plt.subplot(nrows,ncols,figsize=(width,hight))
fig, ax = plt.subplots(1,2,figsize = (10,5))

x  = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y,color='blue')
ax[0].set_title('Line Plot')

ax[1].bar(x,y,color='green')
ax[1].set_title('Bar Chart')

fig.suptitle ('Comarision Of Line And Bar Chart')

plt.tight_layout()

plt.show()
```
