```python
#plt.plot(x,y, color = 'Color name', linestyle = 'line_style',linewidth = value, marker = 'marker symbole', label = 'label name')
import matplotlib.pyplot  as plt

months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months,sales,color = 'blue', linestyle = '--', linewidth = 2, marker = 'o', label = '2025 Sales Data')
plt.xlabel ('Months')
plt.ylabel('Sales per month')
plt.title('Mothly Sales Data Report')
plt.legend(loc='upper left', fontsize=12 ) # upper left - top left corner , lower right- bottum right corner # you can add fontsize also 
#legend work when you set  lable other wise it's not work
plt.grid(color='gray', linestyle = ':', linewidth = 1 )
plt.xlim(1,4) # range of x axis
plt.ylim(0,2000) # range of y axis
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
#plt.bar(x,hight,clor ='color name',width = value,label = 'label name')
import matplotlib.pyplot  as plt

product = ['A','B','C','D']
sales = [1000,800,700,400]

plt.bar(product,sales,color = 'green',label = ' Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()

plt.show()
```

```python
#plt.pie(values,albels=label_list,colors=color_list,autopct='%1.1f%%')
import matplotlib.pyplot  as plt

regions = ['North','South','East','West']
revenue = [3000,4000,2500,1000]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['Red','Blue','green','coral'])
plt.title('Revenue Contribution By Region')
plt.show()
```

```python
#plt.hist(data,bins=number_of_bins,color='colorName',edgecolor='black')
import matplotlib.pyplot  as plt

scores = [45,67,89,56,92,73,75,70,80,87,68,82,66,90]

plt.hist(scores,bins=6,color='red',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Score Distribution of Students')
plt.show()
```

```python
# plt.scatter(x,y,color='color name',marker='marker style',label = 'lable name')
import matplotlib.pyplot  as plt

hours = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]
plt.scatter(hours,exam_scores,color='green',marker='0  ',label='Student Data')
plt.xlabel('Hours Studies')
plt.ylabel('Exam Score')
plt.title('Reletionship Between Time and Exam Score')
plt.grid(True)
plt.legend()
plt.show()

```
